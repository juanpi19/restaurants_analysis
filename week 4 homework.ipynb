```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_file = pd.read_csv('assignment.csv')
df_file
```

```python
# rename the columns
df_file = df_file.rename(columns = {'location': 'RESTAURANT_NAME', 'day': 'DATE', 'number of orders': 'NUM_ORDERS', 'percentage of deliveries': "PERC_DELIVERY"})
df_file.head()
```

```python
df_file.info()
```

```python
df_file.DATE = pd.to_datetime(df_file.DATE)
```

```python
df_file.RESTAURANT_NAME.value_counts()
```

```python
df_file[df_file.RESTAURANT_NAME == 'Bryant Park'].DATE.dt.weekday.value_counts()
```

```python
df_file.NUM_ORDERS.describe()
df_file.NUM_ORDERS.hist(bins = 30)
```

```python
df_file[df_file.NUM_ORDERS == df_file.NUM_ORDERS.max()]
```

```python
df_file[df_file.NUM_ORDERS > 1000].RESTAURANT_NAME.value_counts().plot(kind='bar')
```

```python
df_file[df_file.NUM_ORDERS < 600].RESTAURANT_NAME.value_counts().plot(kind='bar')
```

```python
# plot summer sales for each restaurant
restaurants = ['Bryant Park' , 'Columbia', 'Flatiron', 'Midtown', 'NYU', 'Upper East Side', 'Upper West Side', 'Williamsburg']

for r in restaurants:
    df_file[(df_file.DATE >= '2018-05-01') & 
               (df_file.DATE <= '2018-08-30') & 
                   (df_file.RESTAURANT_NAME == r)].plot(x = 'DATE', y = 'NUM_ORDERS', title = r)
```
